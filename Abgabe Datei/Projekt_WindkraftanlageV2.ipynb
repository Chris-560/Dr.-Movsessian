# Funktion zur Berechnung der Euklidischen Distanz

df_scaled['novelty_index_new'] = np.sqrt(((df_scaled[feature_cov_columns] - healthy_mean) ** 2).sum(axis=1))

```python
def euclidean_distance(column_names):
    """ Berechnet die euklidische Distanz für eine gegebene Menge von Spaltennamen """
    common_columns = list(set(column_names) & set(healthy_mean.index))  # Gemeinsame Spalten finden
    return np.sqrt(((df_scaled[common_columns] - healthy_mean.loc[common_columns]) ** 2).sum(axis=1))
```

# Alle Euklidischen distanzen 

```python
df_scaled_Euklidische_d  = {
    'dist_Cov_1': euclidean_distance(features_Cov_1),
    'dist_Cov_2': euclidean_distance(features_Cov_2),
    'dist_Cov_4': euclidean_distance(features_Cov_4),
    'dist_Cov_5': euclidean_distance(features_Cov_5),
    'dist_Cov_7': euclidean_distance(features_Cov_7),
    'dist_Cov_8': euclidean_distance(features_Cov_8),
    'dist_Cov_10': euclidean_distance(features_Cov_10),
    'dist_Cov_11': euclidean_distance(features_Cov_11),
    'dist_Cov_gesamt': euclidean_distance(feature_cov_columns)
}

distances_df_scaled = pd.DataFrame(df_scaled_Euklidische_d)
```

# Optimalen Schwellenwert berechnen

```python


def plot_combined_analysis(column_name):
    """Plottet die ROC-Kurve mit Schwellenwert und Novelty Index für die ausgewählte Spalte."""
    # ROC-Kurve und Schwellenwerte berechnen
    fpr, tpr, thresholds = roc_curve(distances_df_label[f'label_{column_name}'], 
                                    distances_df_scaled[column_name])
    
    # Youden's J-Index berechnen
    j_scores = tpr - fpr
    roc_auc = auc(fpr, tpr)
    
    # Optimalen Schwellenwert finden
    optimal_idx = np.argmax(j_scores)
    optimal_threshold = thresholds[optimal_idx]
    
    # Figure mit zwei Subplots erstellen
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Linker Plot (ROC-Kurve mit Schwellenwert)
    axes[0].plot(fpr, tpr, color='blue', lw=2, 
                label=f'ROC-Kurve (AUC = {roc_auc:.2f})')
    axes[0].plot([0, 1], [0, 1], color='gray', 
                linestyle='--', lw=2, label='Zufallslinie')
    
    # Optimalen Punkt markieren
    axes[0].scatter(fpr[optimal_idx], tpr[optimal_idx], color='red', 
                   label=f'Optimaler Schwellenwert = {optimal_threshold:.2f}')
    axes[0].text(fpr[optimal_idx] + 0.02, tpr[optimal_idx] - 0.02, 
                 f'Threshold = {optimal_threshold:.2f}\nJ-Index = {j_scores[optimal_idx]:.2f}', 
                 color='red')
    
    axes[0].set_xlabel('Falsch-Positiv-Rate')
    axes[0].set_ylabel('Richtig-Positiv-Rate')
    axes[0].set_title(f'ROC-Kurve für {column_name}')
    axes[0].legend(loc='lower right')
    axes[0].grid(True)
    
    # Rechter Plot (Novelty Index Scatter mit Schwellenwertlinie)
    scatter = axes[1].scatter(range(len(distances_df_scaled)), 
                            distances_df_scaled[column_name],
                            c=distances_df_label[f'label_{column_name}'], 
                            cmap='bwr', 
                            label='Novelty Index')
    
    # Horizontale Schwellenwertlinie hinzufügen
    axes[1].axhline(y=optimal_threshold, color='green', linestyle='--', 
                    label=f'Schwellenwert ({optimal_threshold:.2f})')
    
    axes[1].set_xlabel('Probenindex')
    axes[1].set_ylabel('Novelty Index')
    axes[1].set_title(f'Novelty Index für {column_name}')
    axes[1].legend()
    
    # Layout optimieren
    plt.tight_layout()
    plt.show()
    
    # Metriken ausgeben
    print(f"\nMetriken für {column_name}:")
    print(f"AUC: {roc_auc:.4f}")
    print(f"Optimaler Schwellenwert: {optimal_threshold:.4f}")
    print(f"J-Index am optimalen Schwellenwert: {j_scores[optimal_idx]:.4f}")

# Rest des Codes bleibt gleich
def on_column_change(change):
    """Wird aufgerufen, wenn die ausgewählte Spalte geändert wird."""
    if change['type'] == 'change' and change['name'] == 'value':
        clear_output(wait=True)
        display(column_dropdown)
        plot_combined_analysis(change['new'])

# Verfügbare Spalten
columns = ['dist_Cov_1', 'dist_Cov_2', 'dist_Cov_4', 'dist_Cov_5', 'dist_Cov_7',
          'dist_Cov_8', 'dist_Cov_10', 'dist_Cov_11', 'dist_Cov_gesamt',
          'novelty_index']

# Dropdown-Menü erstellen
column_dropdown = widgets.Dropdown(
    options=columns,
    value=columns[0],
    description='Spalte:',
    style={'description_width': 'initial'}
)

# Widget mit Callback verbinden
column_dropdown.observe(on_column_change, names='value')

# Initial anzeigen
display(column_dropdown)
plot_combined_analysis(columns[0])
```

```python
import seaborn as sns
from scipy import stats

def analyze_operational_conditions(column_name):
    """
    Analysiert die Zusammenhänge zwischen Betriebsbedingungen und dem Novelty Index
    für die ausgewählte Spalte.
    """
    # Berechne den optimalen Schwellenwert
    fpr, tpr, thresholds = roc_curve(distances_df_label[f'label_{column_name}'], 
                                    distances_df_scaled[column_name])
    j_scores = tpr - fpr
    optimal_threshold = thresholds[np.argmax(j_scores)]
    
    # Erstelle binäre Klassifikation basierend auf dem Schwellenwert
    anomaly_detection = (distances_df_scaled[column_name] > optimal_threshold).astype(int)
    
    # Kombiniere relevante Daten in einem DataFrame
    analysis_df = pd.DataFrame({
        'Novelty_Index': distances_df_scaled[column_name],
        'Is_Anomaly': anomaly_detection,
        'WindSpeed': df_scaled['WindSpeed'],
        'Temperature': df_scaled['Temperature'],
        'Wind_Direction': df_scaled['Wind_Direction'],
        'Power_kW': df_scaled['Power_kW'],
        'Generator_Status': df_scaled['Generator_on_off'],
        'Pitch_deg': df_scaled['Pitch_deg']
    })
    
    # Erstelle Subplot-Grid
    fig = plt.figure(figsize=(20, 15))
    gs = fig.add_gridspec(3, 2, hspace=0.3)
    
    # 1. Windgeschwindigkeit vs. Novelty Index
    ax1 = fig.add_subplot(gs[0, 0])
    sns.scatterplot(data=analysis_df, x='WindSpeed', y='Novelty_Index', 
                   hue='Is_Anomaly', palette='coolwarm', ax=ax1)
    ax1.axhline(y=optimal_threshold, color='green', linestyle='--', label='Schwellenwert')
    ax1.set_title('Windgeschwindigkeit vs. Novelty Index')
    
    # 2. Temperatur vs. Novelty Index
    ax2 = fig.add_subplot(gs[0, 1])
    sns.scatterplot(data=analysis_df, x='Temperature', y='Novelty_Index', 
                   hue='Is_Anomaly', palette='coolwarm', ax=ax2)
    ax2.axhline(y=optimal_threshold, color='green', linestyle='--', label='Schwellenwert')
    ax2.set_title('Temperatur vs. Novelty Index')
    
    # 3. Windrichtung vs. Novelty Index (Polar Plot)
    ax3 = fig.add_subplot(gs[1, 0], projection='polar')
    scatter = ax3.scatter(analysis_df['Wind_Direction'], analysis_df['Novelty_Index'], 
                         c=analysis_df['Is_Anomaly'], cmap='coolwarm')
    ax3.set_title('Windrichtung vs. Novelty Index')
    
    # 4. Leistung vs. Novelty Index
    ax4 = fig.add_subplot(gs[1, 1])
    sns.scatterplot(data=analysis_df, x='Power_kW', y='Novelty_Index', 
                   hue='Is_Anomaly', palette='coolwarm', ax=ax4)
    ax4.axhline(y=optimal_threshold, color='green', linestyle='--', label='Schwellenwert')
    ax4.set_title('Leistung vs. Novelty Index')
    
    # 5. Box Plot für Generator Status
    ax5 = fig.add_subplot(gs[2, 0])
    sns.boxplot(data=analysis_df, x='Generator_Status', y='Novelty_Index', ax=ax5)
    ax5.axhline(y=optimal_threshold, color='green', linestyle='--', label='Schwellenwert')
    ax5.set_title('Novelty Index nach Generator Status')
    
    # 6. Pitch Winkel vs. Novelty Index
    ax6 = fig.add_subplot(gs[2, 1])
    sns.scatterplot(data=analysis_df, x='Pitch_deg', y='Novelty_Index', 
                   hue='Is_Anomaly', palette='coolwarm', ax=ax6)
    ax6.axhline(y=optimal_threshold, color='green', linestyle='--', label='Schwellenwert')
    ax6.set_title('Pitch Winkel vs. Novelty Index')
    
    plt.suptitle(f'Betriebsbedingungen Analyse für {column_name}', size=16)
    plt.tight_layout()
    
    # Korrelationsanalyse
    correlation_vars = ['WindSpeed', 'Temperature', 'Power_kW', 'Pitch_deg']
    correlations = {var: stats.pearsonr(analysis_df[var], analysis_df['Novelty_Index']) 
                   for var in correlation_vars}
    
    print("\nKorrelationsanalyse mit Novelty Index:")
    for var, (corr, p_val) in correlations.items():
        print(f"{var}: r={corr:.3f}, p={p_val:.3e}")
    
    # Statistische Tests für verschiedene Betriebszustände
    generator_stats = stats.mannwhitneyu(
        analysis_df[analysis_df['Generator_Status'] == 1]['Novelty_Index'],
        analysis_df[analysis_df['Generator_Status'] == 0]['Novelty_Index']
    )
    print(f"\nMann-Whitney U Test für Generator Status: p={generator_stats.pvalue:.3e}")

# Dropdown-Menü und Callback aktualisieren
def on_column_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        clear_output(wait=True)
        display(column_dropdown)
        analyze_operational_conditions(change['new'])

# Dropdown-Menü erstellen und anzeigen
column_dropdown = widgets.Dropdown(
    options=columns,
    value=columns[0],
    description='Spalte:',
    style={'description_width': 'initial'}
)

column_dropdown.observe(on_column_change, names='value')
display(column_dropdown)
analyze_operational_conditions(columns[0])
```
