```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Daten laden und vorbereiten
data = pd.read_csv(datei_path)
data = data.drop('novelty_index', axis=1) # Zielvariable entfernen
X = data.values
y = np.array([0, 1, 2, 3]) # Zielvariable definieren (muss noch angepasst werden!)

# 📌 Fehlende Werte prüfen
print("\n❓ Fehlende Werte pro Spalte:")
print(data.isnull().sum())

# 📌 Falls es fehlende Werte gibt -> mit Mittelwert füllen (optional)
data.fillna(data.mean(), inplace=True)

# Daten in Trainings- und Testsets aufteilen
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)

# Zielvariable binarisieren
y_bin = label_binarize(y_train, classes=[0, 1, 2, 3])
n_classes = y_bin.shape[1]

# Klassifikator trainieren (Support Vector Machine)
classifier = SVC(kernel='rbf', probability=True)
y_score = classifier.fit(X_train, y_bin).decision_function(X_test)

# ROC-Kurve und AUC für jede Klasse berechnen
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Mikro-durchschnittliche ROC-Kurve und AUC berechnen
fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# ROC-Kurve plotten
plt.figure()
plt.plot(fpr["micro"], tpr["micro"],
         label='Mikro-durchschnittliche ROC-Kurve (Fläche = {0:0.2f})'
               ''.format(roc_auc["micro"]),
         color='deeppink', linestyle=':', linewidth=4)

for i in range(n_classes):
    plt.plot(fpr[i], tpr[i],
             label='ROC-Kurve der Klasse {0} (Fläche = {1:0.2f})'
             ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Falsch-Positive-Rate')
plt.ylabel('Richtig-Positive-Rate')
plt.title('Empfängerkennlinie')
plt.legend(loc="untere rechts")
plt.show()
```
